````python
import requests
from bs4 import BeautifulSoup

# List of URLs to scrape
urls = [
    "https://www.moneycontrol.com/news/business/tata-motors-launches-punch-icng-price-starts-at-rs-7-1-lakh-11098751.html",
    "https://www.moneycontrol.com/news/business/markets/wall-street-rises-as-tesla-soars-on-ai-optimism-11351111.html"
]

data = []

# Function to fetch and parse the content from a URL
def fetch_and_parse(url):
    try:
        response = requests.get(url)
        if response.status_code == 200:
            soup = BeautifulSoup(response.content, 'html.parser')
            # Extracting the article text (this may vary depending on the website structure)
            paragraphs = soup.find_all('p')
            article_text = ' '.join([para.get_text() for para in paragraphs])
            return article_text
        else:
            print(f"Failed to fetch {url}. Status code: {response.status_code}")
            return None
    except Exception as e:
        print(f"An error occurred while fetching {url}: {e}")
        return None

# Fetch content from each URL
for url in urls:
    content = fetch_and_parse(url)
    if content:
        data.append(content)

# Check the number of articles fetched
print(f"Number of articles fetched: {len(data)}")

# Print the content of the first article (if any)
if data:
    print(f"Content of the first article:\n{data[0]}")
else:
    print("No articles fetched.")
````

````python
from langchain.docstore.document import Document
from langchain.text_splitter import RecursiveCharacterTextSplitter

# Assume 'data' contains the raw text fetched from URLs

# Convert fetched content into Document objects
documents = [Document(page_content=content) for content in data]

# Now split the documents
text_splitter = RecursiveCharacterTextSplitter(
    chunk_size=300,
    chunk_overlap=100
)

docs = text_splitter.split_documents(documents)
len(docs)
````

````python
import requests
from bs4 import BeautifulSoup

# Function to fetch and parse content from a URL
def fetch_url_content(url):
    response = requests.get(url)
    if response.status_code == 200:
        soup = BeautifulSoup(response.content, 'html.parser')
        return ' '.join([p.get_text() for p in soup.find_all('p')])
    return None

urls = [
    "https://www.moneycontrol.com/news/business/tata-motors-launches-punch-icng-price-starts-at-rs-7-1-lakh-11098751.html",
    "https://www.moneycontrol.com/news/business/markets/wall-street-rises-as-tesla-soars-on-ai-optimism-11351111.html"
]

data = [fetch_url_content(url) for url in urls]
````

````python
import requests
import pickle
import os
from bs4 import BeautifulSoup


def fetch_url_content(url):
    response = requests.get(url)
    if response.status_code == 200:
        soup = BeautifulSoup(response.content, 'html.parser')
        return ' '.join([p.get_text() for p in soup.find_all('p')])
    return None

urls = [
    "https://www.moneycontrol.com/news/business/tata-motors-launches-punch-icng-price-starts-at-rs-7-1-lakh-11098751.html",
    "https://www.moneycontrol.com/news/business/markets/wall-street-rises-as-tesla-soars-on-ai-optimism-11351111.html"
]

data = [fetch_url_content(url) for url in urls]

````

````python
from langchain.embeddings import HuggingFaceEmbeddings

documents = [Document(page_content=content) for content in data]

text_splitter = RecursiveCharacterTextSplitter(
    chunk_size=300,
    chunk_overlap=100
)

docs = text_splitter.split_documents(documents)

docs[4]


````

````python
embeddings = HuggingFaceEmbeddings(model_name="sentence-transformers/all-MiniLM-L6-v2")
````

````python
from langchain.vectorstores import FAISS
from langchain.docstore.document import Document



vectorindex_hf = FAISS.from_documents(docs, embeddings)

file_path = "vector_index2.pkl"
with open(file_path, "wb") as f:
    pickle.dump(vectorindex_hf, f)

if os.path.exists(file_path):
    with open(file_path, "rb") as f:
        vectorIndex = pickle.load(f)
````

````python
from langchain.text_splitter import RecursiveCharacterTextSplitter
from openai import OpenAI


def get_relevant_chunks(query, index, top_k=3):
    query_embedding = embeddings.embed_query(query)

    relevant_chunks = index.similarity_search_by_vector(query_embedding, k=top_k)

    return [chunk.page_content for chunk in relevant_chunks]

question = "what is tesla up to  "

relevant_chunks = get_relevant_chunks(question, vectorIndex)

combined_content = " ".join(relevant_chunks)
````

````python
from google.colab import userdata
api_key = userdata.get('GROQ_API_KEY')

if api_key:
    print("API Key retrieved successfully!")
else:
    print("API Key not found!")

client = OpenAI(
  base_url="https://api.groq.com/openai/v1",
  api_key=api_key
)

prompt = f"Here is the content: {combined_content}\n\nQuestion: {question}"

response = client.chat.completions.create(
    model="llama-3.1-70b-versatile",
    messages=[
        {"role": "user", "content": prompt}
    ]
)


print(response.choices[0].message.content)
````
